### Data Import

```python
import numpy as np
import scipy.io.wavfile as wav
import math
from scipy.io import wavfile
import matplotlib.pyplot as plt
from google.colab import files
import scipy.io.wavfile as wavfile
from scipy import signal
```

```python
class Import_Data:
  def __init__(self, Path):
        self.Path = Path
        self.sample_rate, self.audio_data = wav.read(self.Path)
        # Audio Process
        if len(self.audio_data.shape) > 1:
            self.audio_data = self.audio_data[:, 0]
        self.audio_array = np.asarray(self.audio_data)
        N = len(self.audio_array)
        nearest_power_of_2 = 2 ** np.ceil(np.log2(N)).astype(int)
        self.audio_array = np.pad(self.audio_array, (0, nearest_power_of_2 - N), 'constant')
        self.duration = len(self.audio_array) / float(self.sample_rate)
  def SampleRate_change_Hz(self,frequecy):
    num_samples = int(self.duration * frequecy)
    audio_data_frequecy = np.interp(np.linspace(0, len(self.audio_array), num_samples, endpoint=False),np.arange(len(self.audio_array)),self.audio_array)
    return audio_data_frequecy
```

```python
I = Import_Data('/content/1.wav')
audio_data_500 = I.SampleRate_change_Hz(500)
audio_data_2000 = I.SampleRate_change_Hz(2000)
audio_data_16000 = I.SampleRate_change_Hz(16000)
```

```python
wavfile.write('audio_data_500.wav', 500, audio_data_500.astype(np.int16))
wavfile.write('audio_data_2000.wav', 2000, audio_data_2000.astype(np.int16))
wavfile.write('audio_data_16000.wav', 16000, audio_data_16000.astype(np.int16))
```

### class FFT

```python
import numpy as np
import numpy as np
import scipy.io.wavfile as wav
import math
from scipy.io import wavfile
import matplotlib.pyplot as plt
```

```python
class FFT_iFFT:

  ## FFT
  def fft(self,x):
    N = len(x)
    if N <= 1:
        return x
    if N % 2 != 0: 
        x = np.append(x, np.zeros(1))
        N += 1
    even = self.fft(x[::2])
    odd = self.fft(x[1::2])
    T = [np.exp(-2j * np.pi * k / N) * odd[k] for k in range(N // 2)]    
    f = [even[k] + T[k] for k in range(N // 2)] + [even[k] - T[k] for k in range(N // 2)]
    return f

  ## Inverse FFT
  def ifft(self,x):
    N = len(x)
    if N <= 1:
        return x
    if N % 2 != 0: 
        x = np.append(x, np.zeros(1))
        N += 1
    even = self.ifft(x[::2])
    odd = self.ifft(x[1::2])
    T = [np.exp(2j * np.pi * k / N) * odd[k] for k in range(N // 2)] 
    return [(1/N)*(even[k] + T[k]) for k in range(N // 2)] + [(1/N)*(even[k] - T[k]) for k in range(N // 2)]

  ## Plot FFT

  def graph(self,fft_array,frequency):
    plt.plot(np.abs(fft_array))
    plt.title('FFT of audio data resampled at {}'.format(frequency))
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude')
    plt.show()
```

```python
F = FFT_iFFT()
fft_500 = F.fft(audio_data_500)
fft_2000 = F.fft(audio_data_2000)
fft_16000 = F.fft(audio_data_16000)
```

```python
F.graph(fft_500,500)
F.graph(fft_2000,2000)
F.graph(fft_16000,16000)
```

##### Test noise reduction algorithm and view steps of algorithm

```python
import IPython
from scipy.io import wavfile
import noisereduce as nr
import soundfile as sf
from noisereduce.generate_noise import band_limited_noise
import matplotlib.pyplot as plt
import urllib.request
import numpy as np
import io
%matplotlib inline
```

## for sample rate 16000

#### Load data

```python
rate_16000, data_16000=wavfile.read("/content/audio_data_16000.wav") 
data_16000 = data_16000
```

```python
IPython.display.Audio(data=data_16000, rate=rate_16000)
```

```python
fig, ax = plt.subplots(figsize=(20,3))
ax.plot(data_16000)
```

#### add noise

```python
noise_len = 2 # seconds
noise = band_limited_noise(min_freq=2000, max_freq = 12000, samples=len(data_16000), samplerate=rate_16000)*10
noise_clip = noise[:rate_16000*noise_len]
audio_clip_band_limited = data_16000+noise
```

```python
fig, ax = plt.subplots(figsize=(20,3))
ax.plot(audio_clip_band_limited)
```

```python
IPython.display.Audio(data=audio_clip_band_limited, rate=rate_16000)
```

#### Stationary remove noise

```python
reduced_noise_16000 = nr.reduce_noise(y = audio_clip_band_limited, sr=rate_16000, n_std_thresh_stationary=1.5,stationary=True)
```

```python
fig, ax = plt.subplots(figsize=(20,3))
ax.plot(reduced_noise_16000)
```

```python
IPython.display.Audio(data=reduced_noise_16000, rate=rate_16000)
```

```python
wavfile.write('audio_filtered_16000.wav', 16000, reduced_noise_16000.astype(np.int16))
```

#### Plot Original and Filtered Audio

```python
# Normalize the audio signals to range from -1 to 1
signal1 = audio_data_16000 / np.max(np.abs(audio_data_16000))
signal2 = reduced_noise_16000 / np.max(np.abs(reduced_noise_16000))
time = np.arange(len(signal1)) / float(16000)# Set the time vector for the audio signals
plt.plot(time, signal1, label='Original Audio')
plt.plot(time, signal2, label='Filtered Audio')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Audio Signals')
plt.legend()
plt.show()
```
